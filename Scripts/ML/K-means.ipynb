鸢尾花是一种多年生草本植物，有块茎或匍匐状根茎。鸢尾花大而美丽，如鸢似蝶，叶片青翠碧绿，观赏价值很高。鸢尾花数据集(iris)是由现代统计学奠基者之一的罗纳德•费雪(Ronald fisher)在1956年的论文"The use of multiple measurement in taxonomic problems"中引入的一个多元数据集。该数据集包括3种鸢尾花，分别是变色鸢尾(Versicolor)、山鸢尾(Setosa)和维吉尼亚鸢尾(Virginica)，每种各50个样本。每个样本测量了4个特征：花萼长度(sepal length)、花萼宽度(sepal width)、花瓣长度(petal length)、花瓣宽度(petal length)，可以通过这些特征预测鸢尾花卉属于3种中的哪一种。
鸢尾花数据集是有标签(label)数据,标签为鸢尾花所属类别，下面将标签去掉，直接通过使用鸢尾花的特征进行聚类。
由于已知存在3种鸢尾花，所以在分类中将k设为3。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

iris = load_iris()  #导入数据
x = iris.data
X = x[:,:2]     #取花萼长度与花萼宽度2个特征
#Y = iris.target #标签数据

#构建模型及预测
kmeans = KMeans(n_clusters=3)   #调用K-Means模型进行3簇聚类
kmeans.fit(X)   #对数据进行学习/训练
y_pred = kmeans.predict(X)  #对数据进行预测
print("预测结果:",y_pred)  #查看预测结果

#模型评估
score = kmeans.inertia_     #计算样本到其最近的聚类中心的平方距离的总和（簇内平方和）
print("簇内平方和为：", score)
```

```python
#画散点图
plt.subplot(1,2,1)  #画子图1
plt.scatter(X[:,0], X[:,1]) #画构建的数据点
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.subplot(1,2,2)  #画子图2
plt.scatter(X[:,0], X[:,1], c=y_pred) #经过聚类后的散点图
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='*', c='r', linewidths=3) #画出中心点
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

鸢尾花有4个特征数据，下面取所有鸢尾花的特征数据进行k均值算法训练。通过k均值聚类后，按顺序输出151个鸢尾花样本数据的分类结果，其中，数字0，1，2号分别对应不同的鸢尾花类别。

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

iris = load_iris()  #导入数据
iris_data = iris.data
Kmeans3 = KMeans(n_clusters=3, random_state=151).fit(iris_data) #采用K均值算法对151个鸢尾花样本进行聚类，最终聚成3类
print(Kmeans3.labels_)  #按序号显示出第1~151个鸢尾花样本数据的分类结果

iris_pred = Kmeans3.predict(iris_data)  #对数据进行预测
print("预测结果:",iris_pred)  #查看预测结果
score3 = Kmeans3.inertia_     #计算样本到其最近的聚类中心的平方距离的总和（簇内平方和）
print("簇内平方和为：", score3)
```
